## 5. Modelling

```python
df.head()
```

```python
# split data into 'x' and 'y'
np.random.seed(42)
x= df.drop(columns='target', axis=1)
y= df['target']

# Split data into train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

Now we got our data split into train and test sets,, it's time to build machine learning model.

we'll train it (find the pattern) on the training test

And we'll test it (use the patterns) on the test set

searching in google: scikit-learn/machine-learning map

we're going to try 3 different machine learning models:
1. Logistic Regression
2. K-Nearest Neighbours classifier
3. Random Forest Classifier(Ensemble Methods) 

```python
# put models in a dictionary
models= {'logistic Regression': LogisticRegression(max_iter=100),
         'KNN': KNeighborsClassifier(),
         'Random Forest':RandomForestClassifier()}

# create a function
def fit_and_score(models, x_train, x_test, y_train, y_test):
    """
    Fits and evalaute given machine learning models
    models: a dict of different machine learning models
    x_train : traning data set (no labels)
    x_test : testing data set (no labels)
    y_train : traning labels
    y_test : test labels 
    
    """
    # set random seed
    np.random.seed(42)
    # make a dictionary to keep model scores
    model_scores= {}
    # Loop through models
    for name, model in models.items():
        # fit the model to the data 
        model.fit(x_train, y_train)
        # Evaluate the model and append the model to the model_scores
        model_scores[name] = model.score(x_test, y_test)
    return model_scores
```

```python
model_scores= fit_and_score(models=models, 
                            x_train= x_train, 
                            x_test= x_test,
                            y_train= y_train,
                            y_test= y_test)
model_scores
```

```python
model_compare= pd.DataFrame(model_scores, index=['accuracy'])
model_compare
```

```python
model_compare.T.plot.bar();
```

### Hyperparameter Turning(By Hand) 

```python
# let's tune KNN

train_scores= []
test_scores= []

# create a list different values for n_neighbors 
neighbors= range(1, 21)

# setup KNN instance 
knn= KNeighborsClassifier()

# loop through differents n_neighbors 
for i in neighbors:
    knn.set_params(n_neighbors=i)

    # Fit the algorithm 
    knn.fit(x_train, y_train)

    # Updates the traning score list 
    train_scores.append(knn.score(x_train, y_train))

    # Updates the test score list 
    test_scores.append(knn.score(x_test, y_test))
```

```python
train_scores
```

```python
test_scores
```

```python
plt.plot(neighbors, train_scores, label='Train score')
plt.plot(neighbors, test_scores, label='Test score')
plt.xticks(np.arange(1,21,1))
plt.xlabel('Number of Neighbors')
plt.ylabel('model score')
plt.legend()

print(f'Maximum KNN score on the test data: {max(test_scores)*100:.2f}')
```

## Hyperparameter Turning with RandomizedSearchCV

we're going to tune:
* LogisticRegression()
* RandomForestClassifier()
  
..... using RandomizedSearchCV
  

```python
# Create Hyperparameter grid for LogisticRegression 
log_reg_grid= {'C': np.logspace(-4, 4, 20), 
               'penalty': ['l1', 'l2'],
               'solver':['liblinear']}

# Create a hyperparameter grid for RandomForestClassifer
rf_grid= {'n_estimators': np.arange(10,1000, 50),
          'max_depth': [None, 3, 5, 10],
          'min_samples_split': np.arange(2, 20, 2),
          'min_samples_leaf': np.arange(1, 20, 2)}
```

Now we have got hyperparaameter grids setup for each of our models, let's tune them using RandomizedSearchCV...

```python
# Tune logisticRegression 
np.random.seed(42)

# setup random hyperparameter search for LogisticRegression 
rs_log_reg= RandomizedSearchCV(estimator=LogisticRegression(),
                               param_distributions=log_reg_grid, 
                               cv=5,
                               n_iter=20,
                               verbose=True)

# Fit random hyperparameter search model for LogisticRegresiion
rs_log_reg.fit(x_train, y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(x_test, y_test)
```

Now we've tuned LogisticRegression(), let;s do the ssame for RandomForestClassifer()...

```python
# setup Random seed
np.random.seed(42)

# setup random hyperparameters search for RandomForestClassifer
rs_rf=RandomizedSearchCV(estimator=RandomForestClassifier(), 
                         param_distributions=rf_grid, 
                         n_iter=20, 
                         cv=5, 
                         verbose=True)

# Fit random hyperparameter search model for RandomForestClassifer
rs_rf.fit(x_train, y_train)
```

```python
# Find the best hyperparameter 
rs_rf.best_params_
```

```python
rs_rf.score(x_test, y_test)
```

```python
# visualizing the confusion matrix
sns.set(font_scale=1.5)

def plot_conf_mat(y_test, y_preds):
    """
    plot a nice looking confusion matrix  using seaborn's Heatmap()
    """
    fig, ax= plt.subplots(figsize=(4,4))
    ax= sns.heatmap(confusion_matrix(y_test, y_preds),
                    annot=True,
                    cbar=False)
    plt.xlabel('True Label')
    plt.ylabel('Predicted label')
    plt.show()
    
plot_conf_mat(y_test, y_preds)
```

NB:
* True positive = model predicts 1 when truth is 1
* False positive = model predicts 1 when truth is 0
* True negative = model predicts 0 when truth is 0
* False negative = model predicts 0 when truth is 1

```python

# import confusion matrix and ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

#Visualizing the confusion matrix

confu_matix= confusion_matrix(y_test, y_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=confu_matix)
# plot the confusion matrix 
disp.plot(cmap='Oranges')
plt.show()
```

### Calculate Evaluation metric using Cross-validation

we're are going to calculate precision, recall, and f1-score of our model using cross-validation and to do so we'll be using 'cross_val_score()'

```python
# check best hyperparameter 
gs_log_reg.best_params_
```

```python
# create a new classifier with best hyperparameter 
clf= LogisticRegression(C= 0.2329951810515372,
                        penalty='l1',
                        solver='liblinear')
```

```python
# import cross_val_score() from  sklearn model_selection 
from sklearn.model_selection import cross_val_score

# Cross-vvalidated accuracy
cv_acc= cross_val_score(clf,
                        x,
                        y,
                        cv=5,
                        scoring='accuracy')
cv_acc=np.mean(cv_acc)
cv_acc
```

```python
# cross-validated precision 
cv_prec= cross_val_score(clf,
                         x,
                         y,
                         cv=5,
                         scoring='precision')
cv_prec= np.mean(cv_prec)
cv_prec
```

```python
# Cross-validated Recall
cv_recall= cross_val_score(clf,
                           x,
                           y,
                           cv=5,
                           scoring='recall')
cv_recall= np.mean(cv_recall)
cv_recall
```

```python
# Cross-validated F1_score
cv_f1= cross_val_score(clf,
                             x,
                             y,
                             cv=5,
                             scoring='f1')
cv_f1= np.mean(cv_f1)
cv_f1
```

```python
# Visualizing our cross_validation metrics
cv_metrics= pd.DataFrame({'Accuracy': cv_acc,
                          'precision': cv_prec,
                          'Recall': cv_recall,
                          'f1_score': cv_f1},
                         index=[0])
cv_metrics.T.plot.bar(title='Cross-validated classification metrics', legend=False);
```
